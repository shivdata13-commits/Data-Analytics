```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset
url = "https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv"
df = pd.read_csv(url)

# Checking sample data
df.head(20)
```

```python
print(df.columns)
```

```python
# Replace '?' and 'Unkown/Invaid with NaN
df.replace("?", np.nan, inplace=True)
df.replace("Unknown/Invalid", np.nan, inplace=True)

# Select relevant columns
df = df[['gender', 'num_procedures']]

# Drop missing values
df.dropna(inplace=True)

# Convert to numeric
df['num_procedures'] = pd.to_numeric(df['num_procedures'])

df.head()
```

```python
# Descriptive statistics
df.groupby('gender')['num_procedures'].describe()
```

```python
df.groupby('gender')['num_procedures'].mean().plot.bar(color=['skyblue', 'salmon'])
plt.title("Average Number of Procedures by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Procedures")
plt.xticks(rotation=0)
plt.show()
```

```python
## Conclusion

## Female patients tend to undergo slightly more procedures on average than male patients.
```
