```python

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df_nypd = pd.read_csv("/content/NYPD_Arrest_Data.csv")

#Display first 10 rows of raw data

df_nypd.head(10)

```

```python
#Data Cleaning
# Drop rows with missing key values
df_nypd = df_nypd.dropna(subset=['ARREST_BORO', 'OFNS_DESC', 'AGE_GROUP'])

# Remove rows where AGE_GROUP is '(null)'
df_nypd = df_nypd[df_nypd['AGE_GROUP'] != '(null)']
```

```python
# Compare Queens (Q) vs Manhattan (M)
comparison_df = df_nypd[df_nypd['ARREST_BORO'].isin(['Q', 'M'])]
```

```python
# Total arrests by borough
arrest_counts = comparison_df['ARREST_BORO'].value_counts()
print("Arrest Counts by Borough:\n", arrest_counts)
```

```python
# Summary statistics
print("\nSummary Statistics:\n", comparison_df.describe())
```

```python
# Bar chart: total arrests
arrest_counts.plot(kind='bar')
plt.title('Number of Arrests: Queens vs Manhattan')
plt.xlabel('Borough')
plt.ylabel('Number of Arrests')
plt.show()
```

```python
# Age group distribution
age_distribution = pd.crosstab(comparison_df['AGE_GROUP'], comparison_df['ARREST_BORO'])
print("\nAge Group Distribution:\n", age_distribution)
```

```python
age_distribution.plot(kind='bar')
plt.title('Number of Arrests By Age_group')
plt.xlabel('Age_group')
plt.ylabel('Number of Arrests')
plt.show()
```

```python
#Conclusion : The results shows Queens has higher number of arrests than Manhattan.
# Age_group 25-44 has higher number of arrests than any other age group.
```
